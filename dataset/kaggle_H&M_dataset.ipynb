<h3>2. Preprocess Dataset</h3>

```python
import pandas as pd 
import numpy as np 
```

```python
meta_data = pd.read_csv('articles.csv')
# Use meta data with article_id starting with "092"
meta_data['article_id'] = meta_data['article_id'].astype(str)
meta_data = meta_data[meta_data['article_id'].str.startswith('92')]
meta_data.head()
```

```python
len(meta_data)
```

```python
# Filter meta data to only have adult wears 
index_name_to_extract = ['Ladieswear', 'Menswear', 'Divided', 'Swimwear']
meta_data = meta_data[meta_data['index_name'].isin(index_name_to_extract)]
len(meta_data)
```

```python
# Filter meta data to have only graments, shoes, accessories, swimwear, nightwear, bag
product_group_name_to_extract = ['Garment Upper body', 'Garment Lower body', 'Shoes', 'Garment Full body', 'Accessories', 'Swimwear', 'Nightwear', 'Bags']
articles_extracted_by_product_name = meta_data[meta_data['product_group_name'].isin(product_group_name_to_extract)]
len(meta_data)
```

<h3>3. Preprocess image dataset </3>

```python
import os 
from PIL import Image 
```

```python
src_image_folder_path = '092'
dst_image_folder_path = 'image_dataset'
for article_id in meta_data['article_id']:
    article_id = '0' + article_id
    # Define file path 
    jpg_file_path = os.path.join(src_image_folder_path, f'{article_id}.jpg')
    png_file_path = os.path.join(dst_image_folder_path, f'{article_id}.png')

    # Convert jpg to png 
    image = Image.open(jpg_file_path)
    image.save(png_file_path, 'PNG')
```
